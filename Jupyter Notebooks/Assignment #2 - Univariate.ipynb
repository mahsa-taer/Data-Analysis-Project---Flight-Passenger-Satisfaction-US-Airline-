```python
df_train = pd.read_csv("Airline_Passenger_Satisfaction_Train_Cleaned.csv")
df_test = pd.read_csv("Airline_Passenger_Satisfaction_Test_Cleaned.csv")

def conv_cat(data):
    data['CustomerType'] = data['CustomerType'].map({'Loyal Customer': 1, 'disloyal Customer': 0})
    data['TypeOfTravel'] = data['TypeOfTravel'].map({'Personal Travel':0, 'Business travel':1})
    data['Class'] = data['Class'].map({'Business':2, 'Eco Plus':1, 'Eco':0})
    data['satisfaction'] = data['satisfaction'].map({'neutral or dissatisfied':0, 'satisfied':1})
    data = pd.get_dummies(data)
    data.drop('id',axis=1,inplace=True)
    data.fillna(0,inplace=True)
    return data
df_train = conv_cat(df_train)
df_test = conv_cat(df_test)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
from math import *


# df_train = pd.read_csv("Airline_Passenger_Satisfaction_Train_Cleaned.csv")
flight_distance = df_train['FlightDistance']

flight_distance_mean = flight_distance.mean()
flight_distance_std = flight_distance.std()
flight_distance_count = len(flight_distance)

print(flight_distance_mean)
print(flight_distance_std)
print(flight_distance_count)

t_value_1 = stats.t.ppf(0.975, 97246)
print(t_value_1)

conf_interval_from = flight_distance_mean - t_value_1 * (flight_distance_std / sqrt(flight_distance_count))
conf_interval_to = flight_distance_mean + t_value_1 * (flight_distance_std / sqrt(flight_distance_count))
print (conf_interval_from, conf_interval_to)
print(t_value_1 * (flight_distance_std / sqrt(flight_distance_count)))
```

<div class="alert alert-block alert-info"><em>The 95% confidence interval for the proportion œÄ of the satisfaction among the
entire population of the airline‚Äôs customers

```python
from scipy import stats
from math import *

cust_satisfaction = df_train['satisfaction']
cust_satisfaction_count = len(cust_satisfaction)
cust_satisfaction_rate = len(cust_satisfaction[cust_satisfaction == 1]) / len(cust_satisfaction)

z_value = stats.norm.ppf(0.975)
print(z_value)

conf_interval_from = cust_satisfaction_rate - z_value * sqrt((cust_satisfaction_rate * (1 - cust_satisfaction_rate))/cust_satisfaction_count)
conf_interval_to = cust_satisfaction_rate + z_value * sqrt((cust_satisfaction_rate * (1 - cust_satisfaction_rate))/cust_satisfaction_count)
print (conf_interval_from, conf_interval_to)

print(z_value * sqrt((cust_satisfaction_rate * (1 - cust_satisfaction_rate))/cust_satisfaction_count))
```

<div class="alert alert-block alert-info"><em>Of 1000 customers who flew with the airline, 100 were satisfied. 
For the following confidence levels, construct a confidence interval for the population proportion
who would be satisfied of the flights. Comment on the relationship between the confidence interval width and the confidence levels.

a. 90%

b. 95%

c. 99%

```python
total_customers = 1000
satisfied = 100
satisfied_rate = satisfied / total_customers

print('Confidence Interval: 90%')

z_value_1 = stats.norm.ppf(0.95)

print(z_value_1)

conf_interval_from = satisfied_rate - z_value_1 * sqrt((satisfied_rate * (1 - satisfied_rate))/total_customers)
conf_interval_to = satisfied_rate + z_value_1 * sqrt((satisfied_rate * (1 - satisfied_rate))/total_customers)
print (conf_interval_from, conf_interval_to)
print(z_value_1 * sqrt((cust_satisfaction_rate * (1 - cust_satisfaction_rate))/cust_satisfaction_count))

#########################################################################

print('\nConfidence Interval: 95%')

z_value_2 = stats.norm.ppf(0.975)

print(z_value_2)

conf_interval_from = satisfied_rate - z_value_2 * sqrt((satisfied_rate * (1 - satisfied_rate))/total_customers)
conf_interval_to = satisfied_rate + z_value_2 * sqrt((satisfied_rate * (1 - satisfied_rate))/total_customers)
print (conf_interval_from, conf_interval_to)
print(z_value_2 * sqrt((cust_satisfaction_rate * (1 - cust_satisfaction_rate))/cust_satisfaction_count))

#########################################################################

print('\nConfidence Interval: 99%')

z_value_3 = stats.norm.ppf(0.995)

print(z_value_3)

conf_interval_from = satisfied_rate - z_value_3 * sqrt((satisfied_rate * (1 - satisfied_rate))/total_customers)
conf_interval_to = satisfied_rate + z_value_3 * sqrt((satisfied_rate * (1 - satisfied_rate))/total_customers)
print (conf_interval_from, conf_interval_to)
print(z_value_3 * sqrt((cust_satisfaction_rate * (1 - cust_satisfaction_rate))/cust_satisfaction_count))
```
